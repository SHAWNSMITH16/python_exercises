```python
def length(unit, quantity):  
    '''
    this function takes length and returns unit conversion
    '''
    if unit == 'cm':
        return (f'{quantity} cm is {round((quantity / 2.54), 2)} in')
    elif unit == 'in':
        return (f'{quantity} in is {round((quantity * 2.54), 2)} cm')
    elif unit == 'miles':
        return (f'{quantity} miles is {round((quantity * 1.6), 2)} km')
    elif unit == 'km':
        return (f'{quantity} km is  {round((quantity * .62), 2)} miles')
    
def mass(unit, quantity):
    '''
    this function takes mass and returns unit conversion
    '''
    if unit == 'lb':
        return (f'{quantity} lb is {round((quantity / 2.2), 2)} kg')
    else: 
        return (f'{quantity} kg is {round((quantity * 2.2), 2)} lb')
    
def volume(unit, quantity):
    '''
    this function takes volume and returns unit conversion
    '''
    if unit == "fl oz":
        return (f'{quantity} fluid ounces is {round((quantity * 29.57), 2)} milliliters')
    else:
        return (f'{quantity} milliliters {round((quantity / 29.57), 2)} fluid ounces')

    
def unit_converter():
    '''
    this function takes input and directs to the appropriate conversion function
    '''
    type_of_meas = input('What type of measurement do you want to convert? length/volume/mass: ')
    if type_of_meas == 'length':
        unit = input('Unit to convert from: \n(cm, in, miles, km) \n')
    elif type_of_meas == 'volume':
        unit = input('Unit to convert from: \n(fl oz or ml) \n')
    elif type_of_meas == 'mass':
        unit = input('Unit to convert from: \n(lb or kg) \n')
        
    quantity = float(input('What quantity?\n'))
    
    if unit in ['cm', 'in', 'miles', 'km']:
        print(f'{length(unit,quantity)}')
    elif unit in ['lb','kg']:
        print(f'{mass(unit, quantity)}')
    elif unit in ['ml', 'fl oz']:
        print(f'{volume(unit, quantity)}')
        

unit_converter()

```
